# Extract contents from html

```python
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
import re

products = []

for page_index, page_html in enumerate(tqdm_notebook(page_sources)):

    soup = BeautifulSoup(page_html, 'html.parser')

    # Find all product containers with class 'eb642d'
    product_elements = soup.find_all(class_="eb642d")

    for el in product_elements:
        try:
            img_tag = el.find("img")
            link_tag = el.find("a")
            no_scipt = el.find("noscript")
            if no_scipt is not None:
                no_scipt = BeautifulSoup(str(no_scipt), 'html.parser')
            
            

            if img_tag and link_tag:
                desc = img_tag.get("alt")
                main_img = img_tag.get("src") # src set is loaded for first few items.
                
                if no_scipt is not None: # if it is not loaded, it is in the noscript tag.
                    main_img = no_scipt.find("img").get("src")

                                    
                hover_img = img_tag.get("data-altimage")
                href = link_tag.get("href")
                article_id = href.split(".")[-2] # neds with productpage.product_id.html, so second last item is id.
                
                image_url = f"https://image.hm.com/{hover_img}" if hover_img else main_img # if there is an alternate/hover image, that is usually the correct image.
                image_url = re.sub(r'\.jpg.*$', '.jpg', image_url) # replace all instances of imwidth


                products.append(
                    {
                        "article_id":article_id,
                        "product_desc": desc,
                        "product_url": href,
                        "image_url": image_url,
                    }
                )
        except Exception as e:
            print(page_index)
            print("Skipped a product due to error:", e)

```

# Download images

```python
import requests
import os
import time
import pandas as pd
from tqdm.notebook import tqdm

# --- Configuration ---
save_directory = f'{os.getcwd()}/images'
download_delay = 0.2  # Adjust this value in seconds
user_agent = 'firefox/1.0'

# Assuming you have a DataFrame called result_df with 'item_image' and 'article_id' columns
for index, article in tqdm(df_unique.iterrows(), total=len(df_unique), desc="Downloading Images"):
    article_id = article['article_id']
    image_url = article.image_url +"?imwidth=500"
    image_filename = os.path.join(save_directory, f'{article_id}.jpg')

    if os.path.exists(image_filename):
        #print(f"Image already exists: {image_filename}. Skipping download.")
        continue
    else:
        try:
            os.makedirs(save_directory, exist_ok=True)
            headers = {'User-Agent': user_agent}
            response = requests.get(image_url, headers=headers)  # Removed stream=True
            response.raise_for_status()
            with open(image_filename, 'wb') as file:
                file.write(response.content)  # Write the entire content at once
            #print(f"Image {index + 1}/{len(result_df)} downloaded successfully and saved to: {image_filename}")
        except requests.exceptions.RequestException as e:
            print(f"Error downloading image for ID {article_id}: {e}")
        except OSError as e:
            print(f"Error creating directory or saving file for ID {article_id}: {e}")
        finally:
            time.sleep(download_delay)

print("Finished processing images.")
```
